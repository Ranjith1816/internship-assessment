```python
import pandas as pd
import sqlite3
```

```python
orders_df = pd.read_csv("/content/orders.csv")

orders_df.head()

orders_df.info()
```

```python
users_df = pd.read_json("/content/users.json")

users_df.head()
users_df.info()

```

```python
conn = sqlite3.connect("/content/restaurants.db")
cursor = conn.cursor()
```

```python
with open("/content/restaurants.sql", "r") as file:
    sql_script = file.read()

# Fix: Drop table if it already exists before creating it
sql_script = "DROP TABLE IF EXISTS restaurants;\n" + sql_script

cursor.executescript(sql_script)
conn.commit()
```

```python
restaurants_df = pd.read_sql_query(
    "SELECT * FROM restaurants",
    conn
)

restaurants_df.head()
restaurants_df.info()

```

```python
orders_users_df = pd.merge(
    orders_df,
    users_df,
    how="left",
    on="user_id"
)
```

```python
final_df = pd.merge(
    orders_users_df,
    restaurants_df,
    how="left",
    on="restaurant_id"
)
```

```python
final_df.head()
final_df.info()

```

```python
final_df.to_csv(
    "/content/final_food_delivery_dataset.csv",
    index=False
)
```
